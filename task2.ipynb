```python
texts = ["""Anaconda is a distribution of the Python and R programming languages for scientific computing, that 
aims to simplify package management and deployment. The distribution includes data-science packages suitable for Windows, 
Linux, and macOS."""]
for text in texts:
	sentences = nltk.sent_tokenize(text)
	for sentence in sentences:
		words = nltk.word_tokenize(sentence)
		print(words)
		tagged = nltk.pos_tag(words)
		print(tagged) 
```
